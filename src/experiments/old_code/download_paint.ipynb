```python
# Cell 1: Imports and setup
import pathlib

import paint.util.paint_mappings as mappings
from paint.data.stac_client import StacClient
from paint.util import set_logger_config

set_logger_config()
```

```python
# Cell 2a: Disable SSL verification (temporary fix)
import ssl
import urllib3

# Disable SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Create unverified SSL context
ssl._create_default_https_context = ssl._create_unverified_context

print("SSL verification disabled (use only for testing)")
```

```python
# Cell 2: Set output directory
output_dir = pathlib.Path.cwd() / "paint_dataset"

print(f"Downloading data to: {output_dir}")
print("=" * 80)
```

```python
# Cell 3: Create STAC client
client = StacClient(output_dir=output_dir)
print("STAC client initialized")
```

```python
# Cell 4: Download tower measurements
print("\n[1/3] Downloading tower measurements...")
client.get_tower_measurements()
print("✓ Tower measurements downloaded")
```

```python
# Cell 5: Download heliostat data for ALL heliostats
print("\n[2/3] Downloading heliostat data for all heliostats...")
print("Collections: properties, calibration, deflectometry")
print("Calibration keys: flux_image, calibration_properties")

client.get_heliostat_data(
    heliostats=None,  # All heliostats
    collections=[
        mappings.SAVE_PROPERTIES.lower(),      # "properties"
        mappings.SAVE_CALIBRATION.lower(),     # "calibration"
        mappings.SAVE_DEFLECTOMETRY.lower(),   # "deflectometry"
    ],
    filtered_calibration_keys=[
        mappings.CALIBRATION_FLUX_IMAGE_KEY,        # "flux_image"
        mappings.CALIBRATION_PROPERTIES_KEY,        # "calibration_properties"
    ],
)
print("✓ Heliostat data downloaded")
```

```python
# Cell 6: Download heliostat metadata for all heliostats
print("\n[3/3] Downloading heliostat metadata for all heliostats...")
print("WARNING: This may take a very long time!")

client.get_heliostat_metadata(heliostats=None)  # All heliostats
print("✓ Heliostat metadata downloaded")
```

```python
# Cell 7: Summary
print("\n" + "=" * 80)
print("Download complete!")
print(f"Data saved to: {output_dir}")
```
